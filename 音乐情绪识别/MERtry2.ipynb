```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import IPython
import librosa
import librosa.display
```

```python
input_music_path="/Users/mac/Desktop/作业/大三下/MER相关论文/音乐情绪识别/MIREX/dramatic/14387.wav"
data, sr = librosa.load(input_music_path)
IPython.display.Audio(data, rate=sr)
```

## 特征提取

```python

result = np.array([])
zcr = np.mean(librosa.feature.zero_crossing_rate(y=data, hop_length=20).T, axis=0)
result = np.hstack((result, zcr))

stft = np.abs(librosa.stft(data))
chroma_stft = np.mean(librosa.feature.chroma_stft(S=stft, sr=sr, n_fft=20, hop_length=20).T, axis=0)
result = np.hstack((result, chroma_stft))

mfcc = np.mean(librosa.feature.mfcc(y=data, sr=sr, n_mfcc=20).T, axis=0)
result = np.hstack((result, mfcc))

rms = np.mean(librosa.feature.rms(y=data, frame_length=100).T, axis=0)
result = np.hstack((result, rms))

mel = np.mean(librosa.feature.melspectrogram(y=data, sr=sr, hop_length=20).T, axis=0)
result = np.hstack((result, mel))
```

```python
X=[]
X.append(result)
X = np.array(X)
dataframe = pd.DataFrame(data = X[0:, 1:],columns = X [0, 1:])
dataframe
```

```python
dataframe.shape
```

```python
X = np.array(dataframe) #X输入数组是将最后一列给去掉后的np数组
X
```

```python
from keras.models import load_model
import numpy as np

# 加载已训练好的模型
model1 = load_model('MusicEmotion.h5')
# 预测测试集
y_pred = model1.predict(X)

# 将预测结果转换为类别标签（如果需要）
y_pred_labels = np.argmax(y_pred, axis=1)

print(y_pred_labels)
print(y_pred)
```
